```python
# Factory Mapping Dictionary

factory_mapping = {
    "Wonka Bar - Nutty Crunch Surprise": "Lot's O' Nuts",
    "Wonka Bar - Fudge Mallows": "Lot's O' Nuts",
    "Wonka Bar -Scrumdiddlyumptious": "Lot's O' Nuts",
    "Wonka Bar - Milk Chocolate": "Wicked Choccy's",
    "Wonka Bar - Triple Dazzle Caramel": "Wicked Choccy's",
    "Laffy Taffy": "Sugar Shack",
    "SweeTARTS": "Sugar Shack",
    "Nerds": "Sugar Shack",
    "Fun Dip": "Sugar Shack",
    "Fizzy Lifting Drinks": "Sugar Shack",
    "Everlasting Gobstopper": "Secret Factory",
    "Hair Toffee": "The Other Factory",
    "Lickable Wallpaper": "Secret Factory",
    "Wonka Gum": "Secret Factory",
    "Kazookles": "The Other Factory"
}

df["Factory"] = df["Product Name"].map(factory_mapping)

df[["Product Name","Factory"]].head()
```

```python
df["Factory"].isnull().sum()
```

```python
features = [
    "Sales",
    "Units",
    "Cost",
    "Gross Profit",
    "Ship Mode",
    "Region",
    "Factory"
]

X = df[features]

y = df["Lead Time"]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

X = X.copy()

X["Ship Mode"] = encoder.fit_transform(X["Ship Mode"])

X["Region"] = encoder.fit_transform(X["Region"])

X["Factory"] = encoder.fit_transform(X["Factory"])

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```python
# Import Linear Regression

from sklearn.linear_model import LinearRegression

# Create the model
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train, y_train)

print("Linear Regression Model Trained Successfully!")
```

```python
# Predict Lead Time

lr_prediction = lr_model.predict(X_test)

print(lr_prediction[:10])
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Calculate Evaluation Metrics

mae = mean_absolute_error(y_test, lr_prediction)

rmse = mean_squared_error(y_test, lr_prediction, squared=False)

r2 = r2_score(y_test, lr_prediction)

print("Linear Regression Performance")
print("MAE :", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(
    random_state=42
)

gb_model.fit(X_train, y_train)

print("Gradient Boosting Model Trained Successfully!")
```

```python
from sklearn.preprocessing import LabelEncoder

le_ship = LabelEncoder()
le_region = LabelEncoder()
le_factory = LabelEncoder()

df["Ship Mode"] = le_ship.fit_transform(df["Ship Mode"])
df["Region"] = le_region.fit_transform(df["Region"])
df["Factory"] = le_factory.fit_transform(df["Factory"])

df.head()
```

```python
X = df[
    [
        "Sales",
        "Units",
        "Cost",
        "Gross Profit",
        "Ship Mode",
        "Region",
        "Factory"
    ]
]

y = df["Lead Time"]

print(X.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(
    random_state=42
)

gb.fit(X_train, y_train)

gb_pred = gb.predict(X_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
import numpy as np

models = {
    "Linear Regression": lr_pred,
    "Random Forest": rf_pred,
    "Gradient Boosting": gb_pred
}

for name, pred in models.items():

    mae = mean_absolute_error(y_test, pred)

    rmse = np.sqrt(mean_squared_error(y_test, pred))

    r2 = r2_score(y_test, pred)

    print("=" * 40)
    print(name)
    print("MAE :", round(mae,2))
    print("RMSE:", round(rmse,2))
    print("R2 :", round(r2,2))
```
